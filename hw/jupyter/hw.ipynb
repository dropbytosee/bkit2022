# Последовательность Фибоначчи

import requests
import matplotlib.pyplot as plt


def get_data(n):
    url = f'http://127.0.0.1:5000/fibonacci?n={str(n)}' 
    request = requests.get(url)
    return list(map(int, request.text.split(' ')))


## Числа Фибоначчи

```python
num = 10
print(f'Первые 10 чисел последовательности Фибоначчи: {get_data(num)}')
```

```python
num = 20
print(f'Первые 20 чисел последовательности Фибоначчи: {get_data(num)}')
```

```python
num = 30
print(f'Первые 30 чисел последовательности Фибоначчи: {get_data(num)}')
```

## Построение графика

```python
num = 10
y_axis = get_data(num)
x_axis = [i for i in range(1, num + 1)]

figure = plt.figure(figsize = (5, 5))
plt.bar(x_axis, y_axis)
plt.xlabel('Номер числа')
plt.ylabel('Значение числа')
plt.title('Первые 10 чисел последовательности Фибоначчи')
plt.show()
```
